```python
"""
Reading file to a Dictionary
class csv.DictReader Creates an object that operates like a regular reader but maps the information in each row to a dict whose keys are given by the optional fieldnames parameter.
"""
import csv
readfile=csv.DictReader(open("cb_sample.csv"))
```

```python
# We can read the element in a dictionary by passing a key to the dictionary in this case key is column 'name' and then looping through dictionary to print all the names
for d in readfile:
    print(d['name'])
```

```python
#Working with Sets in Python
my_set = {1, 2, 3}
print(my_set)


# Python can also deal with a  set of mixed datatypes
my_set = {1.0, "Hello", (1, 2, 3)}
print(my_set)
```

```python
#Adding to a set in Python
my_set.add(4)
print(my_set)

# Adding multiple elements in set
my_set.update([5,6,7])
print(my_set)

#Removing elements from a set
my_set.remove(6)
print(my_set)
```

```python
#Performing Uion operation with Sets
# initialize A and B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# use | operator or union function
print(A | B)
```

```python
#Performing Intersection operation with Sets
# initialize A and B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# use & operator for Set Intersection
print(A & B)
```
